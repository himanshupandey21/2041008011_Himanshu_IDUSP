```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X=[1,2]
Y=[1,3]
plt.scatter(X,Y)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

slope_current=0
intercept_current=0
LR = 0.01
iteration = 1000

for i in range(iteration):
    sum1 = 0
    sum2 = 0
    for j in range(len(X)):
        sum1 += X[j] * (Y[j] - slope_current*X[j] - intercept_current)
        sum2 += Y[j] - slope_current*X[j] - intercept_current
    slope_gradient = -(2/len(X)) * sum1
    intercept_gradient = -(2/len(X)) * sum2
    slope_current -= LR * slope_gradient
    intercept_current -= LR * intercept_gradient
    y_predicted = [slope_current*x + intercept_current for x in X]
    plt.plot(X, y_predicted)
plt.scatter(X, Y)
y_predicted = [slope_current*x + intercept_current for x in X]
plt.plot(X, y_predicted)
plt.show()
```
